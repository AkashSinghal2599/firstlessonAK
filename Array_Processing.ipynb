Here we will create two simple arrays, craete a boolean array which will allow us to input true and false into an array. We will call that boolean array a condition. Which will be used for list comprehension.

```python
#numpy where
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
condition = np.array ([True, True, False, False])
answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
answer
```

Now at the end we made an array where we chose the A value if the condition is true otherwise we chose the B value.

```python
answer2 = np.where(condition,A,B)
answer2
```

Use of numpy where or np where is to prove if the condition you have applied is true or false in a crisper and faster way.
